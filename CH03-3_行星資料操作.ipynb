```python
#練習:

import numpy as np
import pandas as pd

rng = np.random.RandomState(42) #設定亂數種子
ser = pd.Series(rng.rand(5)) #產生5個隨機數字
print(ser)
print("總和", ser.sum())
print("平均數", ser.mean())

data = pd.DataFrame({'colA': rng.rand(5), 'colB': rng.rand(5), 'colC': rng.rand(5)})
print(data)

print("="*60)
print("每一個欄位的平均數:")
print(data.mean())

print("="*60)
print("每一個資料的平均數:")
print(data.mean(axis='columns'))
```

```python
import seaborn as sns

planets = sns.load_dataset("planets")
shapes = planets.shape
print("資料筆數:", shapes[0], "欄位數:", shapes[1])
print(planets.head(10)) #前10筆資料內容
```

```python
import csv

filename = "planets.csv"
with open(filename, 'w',encoding="utf-8",newline="") as fp:
    w = csv.writer(fp, delimiter=",")
    for i in range(shapes[0]):
        rowdata = planets.iloc[i]
        w.writerow(rowdata[j] for j in range(shapes[1]))
```

```python
# 資料集.dropna(axus=0/1, how="any/all", thresh=number, ): 處理缺失值
# axis=0(行)1(col)
# any:只有缺失值出現，刪除該行或是該欄位
# all:該行/該列所有的缺失之都沒有時才刪除
# thresh=number: 該行或是該列，至少有(>=number)的缺失值才刪除

p2 = planets.dropna()
p2_shapes = p2.shape
print(p2_shapes)
```

```python
#幫助資料集做每一個欄位的聚合作用
p2.describe()
```

```python
print("(1)計算項目總數:", p2.distance.count())
print("(2)平均數:", p2.distance.mean())
print("(3)中位數:", p2.distance.median())
print("(4)最小值/最大值:", p2.distance.min(), p2.distance.max())
print("(5)標準差:", p2.distance.std())
print("(6)變異數:", p2.distance.var())
print("(7)平均絕對偏差:", p2.distance.mad())
print("(8)所有項目的乘機:", p2.distance.prod())
print("(9)所有項目的總和:", p2.distance.sum())
```

### 資料集聚合 (Groupby) : split, apply, combine

    1. split: 根據特定的鍵值做分解與分組
    2. apply: 聚合/轉換/過濾
    3. combine: 結合合併

```python
data = pd.DataFrame({'fruit':['A', 'B', 'C', 'D', 'E', 'F', 'G'], 'price':[30, 25, 29, 30, 40, 70, 70]},
                     columns=['fruit','price'], index=['CA', 'NY', 'FL', 'TX', 'KT', 'AZ', 'LA'])
print(data)
```

```python
print(data.groupby('fruit')) #DataFrameGroupBy object
```

```python
data.groupby('price').sum()
```

```python
import seaborn as sns

planets = sns.load_dataset("planets")
shapes = planets.shape
print("資料筆數:", shapes[0], "欄位數:", shapes[1])

#planets = planets.dropna()
```

```python
planets_groupby_method = planets.groupby('method') #產生groupby物件
planets_groupby_method['orbital_period'] #SeriesGroupBy object
planets_groupby_method['orbital_period'].mean()
```

```python
for (method, group) in planets.groupby('method'):
    print(f"{method:30} 資料筆數={group.shape[0]}")
    
```

```python
print(planets.loc[planets.method=='Radial Velocity']) #列出
```

```python
for (y, group) in planets.groupby('year'):
    print(f"{y:10}, 資料筆數={group.shape[0]}")
    print(group)
    print("="*60)
```

```python
for (y,group) in planets.groupby('method')['year']:
    print(f"{y:20}, {group.shape}")
```
