```python
import numpy as np
import pandas as pd

rng = np.random.RandomState(0)
df = pd.DataFrame({'key':['A', 'B', 'C','A','B','C'],'data1':range(6),'data2':rng.randint(0,10,6)}, \
                  columns=['key','data1','data2'])

df
```

```python
print(df.groupby('key').sum())
```

```python
df.groupby('key').aggregate([min,np.median,max,np.mean]) #分層
```

```python
df.groupby('key').aggregate({'data1':min, 'data2':max})
```

```python
# transform(): 轉換

df.groupby('key').transform(lambda x:x-x.mean())
```

```python
import seaborn as sns

planets = sns.load_dataset("planets")
shapes = planets.shape
print("資料筆數:", shapes[0], "欄位數:", shapes[1])

#planets = planets.dropna()
```

```python
decade = 10*(planets['year']//10)
decade = decade.astype(str) + 's'
decade.name = 'decade'
planets.groupby(['method',decade])['number'].sum().unstack()
```

```python
planets.groupby(['method',decade])['number'].sum().unstack().fillna("x")
```

```python
planets.groupby(['method',decade])['number'].sum().unstack().fillna("0")
```

```python
planets.groupby(['method','year'])['number'].sum().unstack().fillna("0")
```
