## Legend(圖像的標籤)

    1. legend entry:(key與label對應)
    2. legend key: marker標記
    3. legend label: key的說明
    4. legend handle: 一個entry圖上對應的物件

```python
import numpy as np
import matplotlib.pyplot as plt

line_up,   = plt.plot([1,2,3],  label='Line 2')
line_down, = plt.plot([3,2,1],  label='Line 1')
plt.legend(handles=[line_up, line_down])
plt.legend(bbox_to_anchor=(1,1), bbox_transform=plt.gcf().transFigure)
plt.show()
```

```python
line1, =plt.plot([1,2,3], label="Line 1", linestyle="--")
line2, =plt.plot([3,2,1], label="Line 2", lw="4")

first_legend = plt.legend(handles=[line1], loc=1)
ax = plt.gca().add_artist(first_legend) #手動把Legend加在目前的Axes上

plt.legend(handles=[line2], loc=4)
plt.show()
```

```python
line1, =plt.plot([1,2,3], label="Line 1", linestyle="--")
line2, =plt.plot([3,2,1], label="Line 2", lw="4")

first_legend = plt.legend(handles=[line1], loc=1)
ax = plt.gca().add_artist(first_legend) #手動把Legend加在目前的Axes上

plt.legend(handles=[line2], loc='lower right')
plt.show()
```

## legend() loc參數的對照:

    0 : best
    1 : upper right
    2 : upper left
    3 : lower right
    4 : lower left
    5 : right
    6 : center left
    7 : center right
    8 : lower center
    9 : upper center
    10: center

```python
z = np.random.randn(10)

red_dot, = plt.plot(z, "ro", markersize=15)
white_cross, = plt.plot(z[:5:], "w+", markersize=15, markeredgewidth=3)
plt.legend([red_dot, (red_dot, white_cross)], ['ProductA', 'ProductB'], loc=4)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
fig = plt.figure()
axes = plt.axes()

x = np.linspace(0, 10, 1000)
#axes.plot(x) #當沒有給y座標陣列時，用index來當作x座標，值為y座標
axes.plot(x,np.sin(x))
```

```python
plt.plot(x, np.sin(x-0), color="blue")
plt.plot(x, np.sin(x-1), color='g')
plt.plot(x, np.sin(x-2), color='0.75') #灰階介於0-1
plt.plot(x, np.sin(x-3), color='#FFDD44')
plt.plot(x, np.sin(x-4), color=(1.0,0.5,0.5))
```

## Line線條樣式


```python
plt.plot(x, x+0, linestyle="solid", color='pink')
plt.plot(x, x+1, linestyle="dashed")
plt.plot(x, x+2, linestyle="dashdot")
plt.plot(x, x+3, linestyle="dotted")
```

```python
t = np.arange(0,5,0.2)
plt.plot(t,t,linestyle="dotted", color='blue')
plt.plot(t,t**2,marker='o',color='r')
plt.plot(t,t**3, 'g^')
plt.plot(t,t**4, 'bs', color='y')
```

's' : square
'D' : Diamond
'^' : 三角形
'<' : 左三角形
