```python
#散布圖

import matplotlib.pyplot as plt
import numpy as np
plt.style.use('seaborn-whitegrid')

x = np.linspace(0,10,30)
y = np.sin(x)

plt.plot(x, y, 'o', color='blue')
```

```python
rng = np.random.RandomState(0) #亂數種子
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker, label='{0}'.format(marker))
    
plt.legend(numpoints=1)
plt.xlim(0,2)
plt.show()
```

```python
plt.plot(x, y, '-x', color='k')
```

```python
plt.plot(x, y, '-.p', color='pink', markersize=15, linewidth=4, markerfacecolor='w',\
         markeredgecolor='gray', markeredgewidth=2)

plt.ylim(-1.2, 1.2)
```

```python
plt.scatter(x, y, marker='o')
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)
sizes  = 1000*rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3, cmap='viridis')
plt.colorbar()
```

```python
#花朵: 特徵(花瓣，萼片)

from sklearn.datasets import load_iris

iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2, s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
```
