```python
# 範例1:

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#解決不能顯示中文
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['Microsoft YaHei']
mpl.rcParams['axes.unicode_minus'] = False

#xlim[座標起始大小,座標結束大小]
#ylim[座標起始大小,座標結束大小]

fig = plt.figure() #建立圖形容器
fig.set_facecolor([0,1,0,0.2])
axes = fig.add_subplot(111) #1 row and 1 col
axes.set(xlim=[0.5,4.5], ylim=[-2,8], title="An Example Axes",ylabel="高度", xlabel="人數")

#呼叫畫出不同圖形: x軸資料, y軸資料, 不同圖形的設定
axes.plot([1,2,3,4],[-1,2,3,7], color="lightblue",linewidth=5)
axes.scatter([0.6,3.8,1.2,2.5],[1.1, 3.7, 5, 7.5], c=[1,2,3,4], marker="^")
plt.show()
```

### 圖形文本函數:

**matplotlub.pylot::**
1. text():     Axess物件任意位置添加TEXT
2. xlabel:     X軸標題
3. ylabel:     Y軸標題
4. title:      給Axes物件添加標題
5. figtext:    給容器添加標題
6. suptitle():
7. anotate():  給Axes物件添加註解，可以加箭頭

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12,9))
fig.suptitle("figure Object SupTitle", fontsize=4, fontweight="bold")
axes = fig.add_subplot(111)
fig.subplots_adjust(top=0.85)
axes.set_title("AXES TITLE")
axes.set_xlabel("X label")
axes.set_ylabel("Y label")

axes.text(3, 8, "boxed text in data coords", style="italic", \
          bbox={'facecolor':'pink', 'alpha':0.5, 'pad':10}, )

axes.text(0.95,0.01,'Colored text in axes coords', \
       verticalalignment='bottom', horizontalalignment='right',\
       transform = axes.transAxes, color='green', fontsize=15)

axes.text(2,6,r'an equation: $E=ac^2$', fontsize=15) # $ --> '數學算式'
axes.plot([2,3], [1,5], 'o')
axes.annotate('annotate point to', xy=(2,1), xytext=(3,4), \
              arrowprops=dict(facecolor='pink', shrink=0.05))

axes.axis([0,10,0,10])
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

fig = plt.figure()
axes = fig.add_subplot(111)

t = np.arange(0.0,5.0,0.01)
s = np.cos(2*np.pi*t)
line = axes.plot(t, s, lw=2) # lw = linewidth

axes.annotate('Local Max:', xy=(3,1), xytext=(0.8, 0.9), textcoords='axes fraction', 
           arrowprops=dict(facecolor='green', shrink=0.05), horizontalalignment='right')
axes.set_ylim(-3, 3)
plt.show
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure(1,figsize=(5,5))
fig.clf() # clf: clean figure

axes = fig.add_subplot(111)
axes.set_aspect(1) #等軸正方形，X軸與Y軸的縮放比例
x1 = -1 + np.random.randn(100)
y1 = -1 + np.random.randn(100)

x2 = -1 + np.random.randn(100)
y2 = -1 + np.random.randn(100)

axes.scatter(x1,y1, color='pink')
axes.scatter(x2,y2, color='blue')

bbox_props = dict(boxstyle='round', alpha=0.8, facecolor='white')
axes.text(-2,-2,"simple A", ha="center", size=20, bbox=bbox_props)
axes.text(2, 2 ,"simple B", ha="center", size=20, bbox=bbox_props)
#ha:herizontalalignment
#va:verticalalighment

bbox_props = dict(boxstyle="rarrow", facecolor=(0.8,0.9,0.8), lw=2,alpha=0.8)
t = axes.text(0, 0, 'Direction', ha='center', va='center', rotation=45, size=15, bbox=bbox_props)
bb = t.get_bbox_patch()
bb.set_boxstyle('rarrow', pad=0.6)

axes.set_xlim(-5,5)
axes.set_ylim(-5,5)
plt.show()
```
