```python
# plot函數

import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0.0, 2.0, 0.01)
y = np.sin(2*np.pi*x)
plt.plot(x,y)

plt.xlabel('time(s)')
plt.ylabel('Voltage(mV)')
plt.title('Voltage VS Time(s)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab   as mlab

x1 = np.linspace(0.0, 5.0)
x2 = np.linspace(0.0, 2.0)

y1 = np.cos(2*np.pi *x1)*np.exp(-x1)
y2 = np.cos(2*np.pi *x2)

#subplot(row, column, index)
plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-') # yellow + dot
plt.title('two plots')
plt.ylabel('data A')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-') #red + 小圓點
plt.xlabel('time(s)')
plt.ylabel('data B')

plt.show()
```

```python
# 長條圖(bar chart)
# bar(x, height, width, bottom(條形x座標), align='center')

n_group = 5
means_men = (20,35,30,35,27)
std_men = (2,3,4,1,2)

means_women = (25,32,34,20,25)
std_women = (3,5,2,3,3)

fig, axes = plt.subplots()
index = (np.arange(n_group))
bar_width = 0.35
opacity = 0.4

s1 = plt.bar(index, means_men, bar_width, alpha=0.4, color='blue', yerr=std_men, label='Men')
s2 = plt.bar(index+bar_width, means_women, bar_width, alpha=0.4, color='red', yerr=std_women, label='Women')

plt.xlabel("Group")
plt.ylabel("Score")
plt.title("Scores by group and gender")
plt.xticks(index+bar_width, ('A', 'B', 'C', 'D', 'E'))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
#圓餅圖: x/sum(x)

labels  = "Frogs", "foxs", "Dogs", "Cats"
sizes   = [15,30,45,10]
colors  = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0,0.2,0,0)

plt.pie(sizes, labels=labels, explode=explode, colors=colors, autopct="%1.1f%%", shadow=True, startangle=90)
plt.show()
```

```python
#圖中有表格

import random

data    = [ [random.randint(30000,300000) for _ in range(5)] for _ in range(5)]
columns = ['Freeze', 'Wind', 'Flood', 'Quake', 'Hail']
rows    = ['%d year' %x for x in (100,50,20,10,5)]
print(rows)

values = np.arange(0,2500,500)
values_increment = 100

colors = plt.cm.BuPu(np.linspace(0,0.5,len(columns))) #產生五種階層顏色
n_rows = len(data)

index = np.arange(len(columns)) + 0.3
bar_width = 0.4

y_offset = np.array([0.0] * len(columns))
cell_text = []
for row in range(n_rows):
    plt.bar(index, data[row], bar_width, bottom=y_offset, color=colors[row])
    y_offset = y_offset + data[row]
    cell_text.append(['%1.1f' % (x/1000.0) for x in y_offset])

colors = colors[::-1]
cell_text.reverse()

the_table = plt.table(cell_text, rowLabels=rows, rowColours=colors, colLabels=columns, loc='bottom')
plt.subplots_adjust(left=0.2, bottom=-0.1)

plt.ylabel("Loss")
plt.xticks([])
plt.title("Loss by Disaster")
```
